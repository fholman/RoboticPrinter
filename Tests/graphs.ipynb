```python
import math

# to calculate time required to print off sheet size

gear_ratio = 20
delay_between_steps = 3 # to include pulse width time as well as time between pulses
pixel_width = 797
pixel_height = 1124
time_per_pixel = 0.008 # in ms - to include both time on and time off of pulse
time_after_each_12 = 0.8
number_of_times_per_ink_drop = 2

step_per_pixel = 3.125 / 12

wheel_circumference = math.pi * 35
distance_per_step = ( ( 18 / gear_ratio ) / 360 ) * wheel_circumference
number_of_steps_per_pixel = round(step_per_pixel / distance_per_step)

total_pixels_time = pixel_width * pixel_height * time_per_pixel * number_of_times_per_ink_drop

if (pixel_height % 12 == 0) :
    number_of_swathes = pixel_width * (pixel_height / 12)
else :
    number_of_swathes = pixel_width * (pixel_height // 12) + pixel_width

additional_time_per_swathe = number_of_swathes * time_after_each_12 * number_of_times_per_ink_drop

vertical_steps = (pixel_height - 12) // 12
if (pixel_height % 12 != 0) :
    vertical_steps += 1
vertical_steps *= 12

print_rows_motor_time = number_of_swathes * delay_between_steps * number_of_steps_per_pixel + vertical_steps * delay_between_steps * number_of_steps_per_pixel

print("time : " + str(total_pixels_time + print_rows_motor_time + additional_time_per_swathe))
print("RPM : " + str((1 / (20 * delay_between_steps)) * 60000))
```
